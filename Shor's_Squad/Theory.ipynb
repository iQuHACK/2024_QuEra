```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
t_f = 0.3
omega = 1.5
r = np.linspace(0, 100, 100)  # range of delta_f values
delta_f = r*t_f


# Calculate the function
function_values = omega**2 / (2 * delta_f) * np.arctanh(delta_f / np.sqrt(omega**2 + delta_f**2))

# Plot the graph
plt.plot(delta_f, function_values)
plt.xlabel('delta_f')
plt.ylabel('Prob.')
plt.title('Switching prob.')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hyp2f1

# Define the expression
def my_function(omega, delta_f, t_f, c6):
    r= delta_f/t_f
    return 0.25* 1/(omega**2 *t_f / (delta_f) * np.arctanh(delta_f / np.sqrt(omega**2 + delta_f**2))) * (omega**2 * t_f * (c6/np.sqrt(omega**2 + r**2*t_f**2))**(7/6) * ((1 + (r**2*t_f**2)/omega**2)**(7/12) *
            hyp2f1(1/2, 7/12, 3/2, -(r**2*t_f**2)/omega**2))) / c6

# Define the parameters
omega = np.linspace(0.5, 10, 100)
t_f = 1
delta_f = np.linspace(1, 20, 100)
c6 = 862690

omega_mesh, delta_f_mesh = np.meshgrid(omega, delta_f)
# Calculate y values
Blockade_radius = my_function(omega_mesh, delta_f_mesh, t_f, c6)

print(my_function(4, 13, 1, c6)*4/4.5)
print(my_function(2.1, 6.8, 1, c6))


# Plot the function
plt.contourf(omega_mesh, delta_f_mesh, Blockade_radius, cmap='viridis', levels =20)
plt.colorbar(label='Re-normalized Rydberg blockade')
plt.title('Effective Total Rydberg Blockade of Omega, Delta System')
plt.xlabel('Omega')
plt.ylabel('Delta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
omega = 2.5
C_6 = 862690
delta_f_values = np.linspace(0, 30, 100)  # range of delta_f values
t_values = np.linspace(0, 3, 100)  # range of t values

# Create meshgrid for delta_f and t
delta_f_mesh, t_mesh = np.meshgrid(delta_f_values, t_values)

# Calculate tau values
tau_mesh = t_mesh  # Same range as t

# Calculate R_b values
R_b_mesh = (C_6 / np.sqrt(omega**2 + (delta_f_mesh * tau_mesh / t_mesh)**2))**(1/6)

# Calculate the function values for coloring
function_values = t_mesh * omega**2 / (2 * delta_f_mesh) * np.arctanh(delta_f_mesh / np.sqrt(omega**2 + delta_f_mesh**2))

# Plot the 3D colormap
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(delta_f_mesh, tau_mesh, R_b_mesh, cmap='viridis', facecolors=plt.cm.viridis(function_values), vmin=np.min(function_values), vmax=np.max(function_values), antialiased=False)
ax.set_xlabel('Delta_f')
ax.set_ylabel('Tau')
ax.set_zlabel('R_b')
ax.set_title('3D Colormap of Switching Probability and R_b')
fig.colorbar(surf, label='Switching Probability')
plt.show()
```
