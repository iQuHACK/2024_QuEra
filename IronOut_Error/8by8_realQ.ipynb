```python
from bloqade import rydberg_h, piecewise_linear, piecewise_constant, waveform, cast, save, load
from bloqade.atom_arrangement import ListOfLocations, Lieb, Square, Chain, Honeycomb, Kagome, Triangular, Rectangular
from bokeh.io import output_notebook # to plot "show()" on the notebook, without opening a browser
import os
import matplotlib.pyplot as plt
import numpy as np

output_notebook()
```

```python
#geometry
rng = np.random.default_rng(1234)
atom_pos=Square(8, lattice_spacing=4).apply_defect_density(0.3, rng=rng).remove_vacant_sites()

#dynamics
d_max = 0.24668
o_max = 0.08333


durations_delta = [.25,3.5,.25]
durations_omega = [0.15, 1.5, .15, 1.85, 0.15]
delta_MHz=[-d_max,-d_max,d_max,d_max]
omega_MHz= [0,o_max,o_max, o_max, o_max, 0]

Delta = piecewise_linear(durations_delta,[x*2*np.pi for x in delta_MHz])
Omega = piecewise_linear(durations_omega,[x*2*np.pi for x in omega_MHz])
program = rydberg_h(atom_pos, detuning= Delta, amplitude=Omega, phase=None)
program.parse_register().show()
program.parse_sequence().show()
```

```python
d_max = 0.24668
o_max = 0.08333
#o_max = 2.5
durations_delta = [.25,3.5,.25]
durations_omega = [0.15, 1.5, .15, 1.85, 0.15]
delta_MHz=[-d_max,-d_max,d_max,d_max]
omega_MHz= [0,o_max,o_max, o_max, o_max, 0]
Delta = piecewise_linear(durations_delta,[x*2*np.pi for x in delta_MHz])
Omega = piecewise_linear(durations_omega,[x*2*np.pi for x in omega_MHz])
program = rydberg_h(atom_pos, detuning= Delta, amplitude=Omega, phase=None)
program.parse_register().show()
program.parse_sequence().show()
```

```python
output_Q = program.braket.aquila().run_async(shots=100)
```
